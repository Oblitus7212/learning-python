```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)

# Normal distributed x and y vector with mean 0 and standard deviation 1
x = np.random.normal(0, 1, 500)
#y = np.random.normal(0, 1, 500)
y = x*.9
X = np.vstack((x, y)).T

plt.scatter(X[:, 0], X[:, 1])
plt.title('Generated Data')
plt.axis('equal');
```

```python
# Covariance
def cov(x, y):
    xbar, ybar = x.mean(), y.mean()
    return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

# Covariance matrix
def cov_mat(X):
    return np.array([[cov(X[0], X[0]), cov(X[0], X[1])], \
                     [cov(X[1], X[0]), cov(X[1], X[1])]])

# Calculate covariance matrix 
cov_mat(X.T) # (or with np.cov(X.T))
```

```python
import numpy as np
x1 = np.array([-2,1]).T
x2 = np.array([-3,5]).T
x3 = np.array([-4,3]).T
x4 = np.array([-5,6]).T
x5 = np.array([-6,7]).T
x6 = np.array([-7,8]).T
x_ = (x1+x2+x3+x4+x5+x6)/6
print('Mean vector is ', x_)
```

```python
dx1 = x1 - x_
dx2 = x2 - x_
dx3 = x3 - x_
dx4 = x4 - x_
dx5 = x5 - x_
dx6 = x6 - x_
```

```python
import numpy as np
adx1 = np.array([dx1])
adx2 = np.array([dx2])
adx3 = np.array([dx3])
adx4 = np.array([dx4])
adx5 = np.array([dx5])
adx6 = np.array([dx6])
m1 = adx1 * adx1.T
m2 = adx2 * adx2.T
m3 = adx3 * adx3.T
m4 = adx4 * adx4.T
m5 = adx5 * adx5.T
m6 = adx6 * adx6.T
Cov = (m1+m2+m3+m4+m5+m6)/6
print('Covariance Matrix = ', Cov)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = [x1[0],x2[0],x3[0],x4[0],x5[0],x6[0]]
y = [x1[1],x2[1],x3[1],x4[1],x5[1],x6[1]]
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

eigvals, eigvecs = np.linalg.eig(Cov)
u = eigvecs[:,1] *-2
lamda = eigvals[1]

Covu = Cov @ u
lamda_u = lamda * u

x_position = [0]   # vector x and b start at origin
y_position = [0]

x_direction =[lamda_u[0]] # List of Endpoint of Arrow for x 
y_direction = [lamda_u[1]]

plt.quiver(x_position, y_position, x_direction, y_direction, angles='xy',scale_units='xy', scale=1,color=['r','b'])
plt.plot(x,y, 'o', color='black')
plt.xlim(-10,1)
plt.ylim(-7,15)

plt.grid(b=True, which='major') #<-- plot grid lines
print('u = ', u); print('lamda *u = ', lamda_u)
```
